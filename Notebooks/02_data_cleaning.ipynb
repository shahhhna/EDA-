### Outlier Detection Method
Outliers in numerical variables were identified using the Interquartile Range (IQR) method.  
According to this approach, observations falling below  
**Q1 − 1.5 × IQR** or above **Q3 + 1.5 × IQR** were flagged as potential outliers.

### 1. Data Type Optimization
* **Categorical Casting:** Customer demographic and account attributes such as `Gender`, `Education_Level`, `Marital_Status`, `Income_Category`, and `Card_Category` were converted from `Object` to `category`. This improves memory efficiency and prepares the data for encoding during modeling.
* **Target Variable Preparation:** The churn indicator `Attrition_Flag` was explicitly cast to a categorical type to clearly distinguish between existing and attrited customers.
* **Identifier Standardization:** The unique customer identifier `CLIENTNUM` was treated strictly as an identifier and excluded from analytical computations to prevent unintended numerical operations.

### 2. Resolution of Missing Values (Nulls)
* **Data Completeness Validation:** A comprehensive missing value audit confirmed that the dataset contains no null values across all features.
* **Imputation Decision:** Since no missing values were detected, no imputation or record removal was required, ensuring full retention of customer information for analysis.

### 3. Text & Feature Standardization
* **Categorical Normalization:** Whitespace trimming and consistent capitalization were applied across all categorical variables to eliminate duplicate category representations caused by formatting inconsistencies.
* **Income Band Harmonization:** The `Income_Category` feature contained mixed textual formats (e.g., "Less than $40K", "40K–60K"). These were standardized into uniform income bands (e.g., `<$40K`, `$40K–$60K`) to improve interpretability.
* **Retention of Informational Categories:** Values labeled as `Unknown` were preserved, as they represent meaningful business segments rather than missing data.

### 4. Outlier Detection & Treatment
* **Statistical Identification:** Outliers in numerical variables were identified using the Interquartile Range (IQR) method, flagging values outside Q1 − 1.5×IQR and Q3 + 1.5×IQR.
* **Behavioral Interpretation:** Detected outliers in features such as `Credit_Limit` and `Total_Trans_Amt` represent high-value or highly active customers and therefore reflect legitimate banking behavior.
* **Treatment Strategy:** Outliers were retained to preserve real-world variability. Logarithmic transformations were applied to highly skewed numerical features to stabilize distributions and support robust analysis.

### 5. Numeric Precision Adjustment
* **Rounding of Derived Metrics:** Ratio-based and percentage change features, including `Total_Amt_Chng_Q4_Q1`, `Total_Ct_Chng_Q4_Q1`, and `Avg_Utilization_Ratio`, were rounded to two decimal places to improve readability and interpretability.
* **Preservation of Raw Values:** Raw financial and count-based variables were intentionally left unrounded to prevent information loss and maintain data accuracy.
